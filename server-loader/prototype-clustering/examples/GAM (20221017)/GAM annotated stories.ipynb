```python
import pandas as pd
import numpy as np
import json
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    New interaction data (annotated-stories.json)
#--------------------------------------------------------------------------------------------------------------------------
```

```python
annotated_stories_df = pd.read_json('data/provided/annotated-stories.json')
annotated_stories_df
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Prepare parts data
#--------------------------------------------------------------------------------------------------------------------------
```

```python
# Explode parts into their elements
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.explode.html
df = annotated_stories_df.explode('parts')
df = df.reset_index()
df
```

```python
# Extract elements from parts
# https://stackoverflow.com/questions/54344114/expand-pandas-dataframe-column-of-dict-into-dataframe-columns
df2 = pd.json_normalize(df["parts"], max_level = 0)
#df2 = pd.json_normalize(df2['multimediaData'], max_level = 0)
df2
```

```python
df3 = pd.concat([df,df2], axis=1, join='inner')
df3
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Group by username
#--------------------------------------------------------------------------------------------------------------------------
```

```python
# First group interaction contributions by userName
df4 = df3.groupby("authorUsername").agg(list)
df4 = df4.reset_index()
df4
```

```python
df5 = df4[['authorUsername', 'artworkId', 'emotions', 'multimediaData']]
df5 = df5.rename(columns = {'authorUsername':'userName'})
df5
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Combine with user information ("relationship_with_arts", "relationship_with_museums")
#--------------------------------------------------------------------------------------------------------------------------
```

```python
users = pd.read_json('data/GAMGame_users_RN_UNITO.json')
users
```

```python
# https://stackoverflow.com/questions/18792918/combine-two-pandas-data-frames-join-on-a-common-column
# Now merge with user data
user_interactions = pd.merge(df5, users, on='userName', how='left')
user_interactions
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html
df = user_interactions[['userName', 'artworkId', 'emotions', 'multimediaData', 'relationship_with_arts', 'relationship_with_museums']]
values = {'relationship_with_arts': '', 'relationship_with_museums': ''}
df = df.fillna(value=values)
df
```

```python
with open("data/processed/annotated-stories.json", "w") as outfile:
    json.dump(df.to_dict('records'), outfile, indent=4)
```
