```python
# Filter columns
artworks2 = artworks[['_id','title','unito_users','web_users']]
artworks2
```

```python
# Get unito_users / web_users
artworks_users = artworks2.copy()
artworks_users = artworks_users.explode('unito_users')
#artworks_users = artworks_users.explode('web_users') 
artworks_users
```

```python
# Extract unito_user_id, sophia_extracted emotions, sophia_extracted_sentiment
# https://stackoverflow.com/questions/63971055/how-to-convert-a-column-of-dictionaries-to-separate-columns-in-pandas
# max_level is to have it stop at the first level of keys (instead of spliting sophia_emotions into columns)
artworks_emotions = artworks_users.copy()
artworks_emotions = pd.json_normalize(artworks_users.unito_users, max_level=0)
artworks_emotions
```

```python
artworks_emotions4 = artworks_emotions.loc[:, artworks_emotions.columns.str.startswith('sophia_extracted_emotions.')]
artworks_emotions4
```

```python
artworks_users2 = artworks_users.reset_index(drop=True)
artworks_users2
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html
# https://www.geeksforgeeks.org/how-to-combine-two-dataframe-in-python-pandas/
# combine both
artworks_unito_emotions = pd.concat([artworks_users2, artworks_emotions], axis = 1, join='inner')
artworks_unito_emotions
```

```python
# https://www.statology.org/pandas-drop-rows-with-condition/
# https://stackoverflow.com/questions/34162625/remove-rows-with-empty-lists-from-pandas-data-frame
artworks_unito_emotions2 = artworks_unito_emotions[['_id','title','unito_user_id','sophia_extracted_emotions']]
artworks_unito_emotions2 = artworks_unito_emotions2[artworks_unito_emotions2['sophia_extracted_emotions'].map(lambda d: len(d)) > 0]
artworks_unito_emotions2
```
