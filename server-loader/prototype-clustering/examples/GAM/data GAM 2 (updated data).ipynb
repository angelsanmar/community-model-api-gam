## Data Analysis - interactions

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Prepare interaction data
#--------------------------------------------------------------------------------------------------------------------------
```

```python
# Split parts array (the result is the number of interactions: emotions, sentiments)
# Number of interactions: 482
df = interactions.explode('parts')
df = df.reset_index()
df
```

```python
# Obtain each element in parts
df2_a = pd.json_normalize(df["parts"], max_level = 0)
df2_a2 = df2_a[['artworkId']]
df2_a2
```

```python
# Obtain each element in multimediaData
df2_b = pd.json_normalize(df2_a["multimediaData"], max_level = 0)
df2_b
```

```python
# Obtain each element in analysis
df2_c = pd.json_normalize(df2_b["analysis"], max_level = 1)
df2_c
```

```python
# Obtain each element in answersToTemplates
df2_d = pd.json_normalize(df2_b["answersToTemplates"], max_level = 0)
df2_d
```

```python
# Create interaction DataFrame with user, artwork and emotions
interactions_df = df.copy()
interactions_df = interactions_df.drop(['parts'], axis=1)
interactions_df
```

```python
# artwork id
interactions_df2 = pd.concat([interactions_df,df2_a2], axis=1, join='inner')
# it makes me feel (text)
interactions_df2 = pd.concat([interactions_df2,df2_d], axis=1, join='inner')
# it makes me feel (emotions)
interactions_df2 = pd.concat([interactions_df2,df2_c], axis=1, join='inner')
interactions_df2
```

```python
# Replace NaN values with {}
columns = ['itMakesMeFeel', 'itMakesMeThinkAbout', 'itRemindsMeOf']
interactionLabels = ['emotions', 'sentiment']
for column in columns:
    for interactionLabel in interactionLabels:
        column2 = column + '.' + interactionLabel
        interactions_df2[column2] = interactions_df2[column2].apply(lambda x: {} if x != x else x)
```

```python
interactions_df2.columns
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Analyze prepared data (interactions_df2)
#--------------------------------------------------------------------------------------------------------------------------
```

```python
df = interactions_df2.copy()
```

```python
print("total interactions: " + str(len(interactions_df2)))
print("\n")

useful_df = []

# Interactions with emotions
columns = ['itMakesMeFeel.emotions', 'itMakesMeThinkAbout.emotions', 'itRemindsMeOf.emotions']
for column in columns:
    result = df.loc[(df[column].str.len() != 0) & (df[column] == df[column]) & (df[column] != {"False": 1.0}) ]
    result = result[[column]]
    
    useful_df.append(result)
    
    print("useful emotion data - " + str(column))
    print(len(result))
    print(result[[column]])
    print("\n") 
    
# Combined one
result_combined_df = pd.concat(useful_df, axis=1, join='outer')
print("useful emotion data (any emotion column is not empty): " + str(columns))
print(len(result_combined_df))
print(result_combined_df[columns])
print("\n")
```

```python
print("total interactions: " + str(len(interactions_df2)))
print("\n")

useful_df = []

# Interactions with emotions
columns = ['itMakesMeFeel.sentiment', 'itMakesMeThinkAbout.sentiment', 'itRemindsMeOf.sentiment']
for column in columns:
    result = df.loc[(df[column].str.len() != 0) & (df[column] == df[column]) & (df[column] != {"False": 1.0}) ]
    result = result[[column]]
    
    useful_df.append(result)
    
    print("useful sentiment data - " + str(column))
    print(len(result))
    print(result[[column]])
    print("\n") 
    
# Combined one
result_combined_df = pd.concat(useful_df, axis=1, join='outer')
print("useful sentiment data (any sentiment column is not empty): " + str(columns))
print(len(result_combined_df))
print(result_combined_df[columns])
print("\n")
```

```python
artworks_df2 = artworks.copy()
artworks_df2['_id'] = artworks_df2['@id'].astype(str)

interactions_analysis = interactions_df2.copy() 
interactions_analysis['artworkId'] = interactions_analysis['artworkId'].astype(str)
```

```python
rn_interactions_merged_outer = pd.merge(artworks_df2, interactions_analysis, left_on='_id', right_on='artworkId', how='outer')
rn_interactions_merged_outer
```

```python
# Artworks not belonging to GAM_Catalogue_Plus with which the users interact in RN_stories
artworksNotInCataloguePlus = rn_interactions_merged_outer.loc[rn_interactions_merged_outer['_id_x'] != rn_interactions_merged_outer['artworkId']]
artworksNotInCataloguePlus = artworksNotInCataloguePlus.dropna(subset=['artworkId'])
artworksNotInCataloguePlus['artworkId'] = artworksNotInCataloguePlus['artworkId'].astype(str)

artworksNotInCatalogue = pd.unique(artworksNotInCataloguePlus['artworkId'])
print(len(artworksNotInCatalogue))
print(artworksNotInCatalogue)
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Interactions with an artwork from GAM_Catalogue_plus (https://spice.kmi.open.ac.uk/dataset/details/113)
#--------------------------------------------------------------------------------------------------------------------------
```

```python
rn_interactions_common = pd.merge(artworks_df2, interactions_analysis, left_on='_id', right_on='artworkId', how='inner')
rn_interactions_common
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Interactions without proper interaction data (after removing interactions without valid artworks)
#--------------------------------------------------------------------------------------------------------------------------
```

```python
print("total interactions: " + str(len(rn_interactions_common)))
print("\n")

# Interactions without emotions
columns = ['itMakesMeFeel.emotions', 'itMakesMeThinkAbout.emotions', 'itRemindsMeOf.emotions']
for column in columns:
    result = rn_interactions_common.loc[rn_interactions_common[column].str.len() == 0 ]
    
    print("empty emotion data - " + str(column))
    print(len(result))
    print("\n")  
```

```python
print("total interactions: " + str(len(rn_interactions_common)))
print("\n")

# Interactions without emotions
columns = ['itMakesMeFeel.sentiment', 'itMakesMeThinkAbout.sentiment', 'itRemindsMeOf.sentiment']
for column in columns:
    result = rn_interactions_common.loc[rn_interactions_common[column].str.len() == 0 ]
    
    print("empty emotion data - " + str(column))
    print(len(result))
    print("\n")  
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Combine interactions with user data
#--------------------------------------------------------------------------------------------------------------------------
```

```python
interactions_grouped = rn_interactions_common.groupby("authorUsername").agg(list)
interactions_grouped
```

```python
rn_user_interactions = pd.merge(deaf_users, interactions_grouped, on='authorUsername', how='right')
rn_user_interactions.dropna(subset = ['artworkId'], inplace = True)
rn_user_interactions
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Users without interactions with valid interaction data
#--------------------------------------------------------------------------------------------------------------------------
```

```python
df = rn_user_interactions.copy()
```

```python
print("total users: " + str(len(df)))
print("\n")

useful_df = []


# Interactions without emotions
columns = ['itMakesMeFeel.emotions', 'itMakesMeThinkAbout.emotions', 'itRemindsMeOf.emotions']
for column in columns:
    df2 = df.explode(column)
    result = df2[(df2[column].str.len() != 0) & (df2[column] == df2[column]) & (df[column] != {"False": 1.0}) ]
    result2 = result.groupby("authorUsername").agg(list)
    
    useful_df.append(result2)
        
    print("useful emotion data - " + str(column))
    print(len(result2))
    print(result2[[column]])
    print("\n")
    
# Combined one
result_combined_df = pd.concat(useful_df, axis=1, join='outer')
print("users with useful emotion data (any emotion column is not empty): " + str(columns))
print(len(result_combined_df))
#print(result_combined_df[columns])
print("\n")
```
