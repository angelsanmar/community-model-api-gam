**Stories datasets with emotions**

Stories datasets for DMH: Narratives 304 stories (first 213 + second 91) stories from VR Museum (Helsinki) about design objects from workshop participants, classified by DEGARI (UNITO), Semantic Annotator (H-FARM), Putchik and SenticNET7 API.

Dataset 1 - 91 stories
- Name: DMH_91Stories_DEGARI_SemanticAnnotator
- UUID:  1396cecb-c11b-4f73-9b2c-7d4b01891c44
- Link:   https://spice.kmi.open.ac.uk/dataset/details/107

Dataset 2 - 213 stories
- Name:  DMH_213Stories_DEGARI_SemanticAnnotator
- UUID:   46ed6774-8af7-468c-b6a8-a0544d0b1ad2
- Link:     https://spice.kmi.open.ac.uk/dataset/details/104

In particular, each json-story is described by the following characteristics

    DEGARI emotions (emotions extracted by our system DEGARI-UNITO)
    Semantic Annotator
    Plutchik 8-emotions {anger, anticipation, disgust, fear, joy, sadness, surprise, trust} and its likelihood 
    concept parsing
    polarity classification
    intensity ranking
    emotion recognition
    aspect extraction
    personality prediction (OCEAN)
    sarcasm identification
    depression categorization
    toxicity spotting

## Connections between stories and objects (DMH - Pop_up_VR_Museum - Objects classification - English) 

DMH - Pop_up_VR_Museum - Written Stories
- **Object number**: links to **Object Number** in Objects classification.

DMH_91Stories_DEGARI_SemanticAnnotator.json
- Nothing linking to objects.
    
DMH_213Stories_DEGARI_SemanticAnnotator.json
- **Artefact**: links to **Object name** in Objects classification. 
- **Artefact_ID**: links to **Object number** in Objects classification. 

DMH_Stories collected from users / DMH_Narratives 123 stories - DEGARI Classification
- **artefact/Artefact**: links to **Object name** in Objects classification.
- **artefact_id/Artefact_ID**: links to **Object number** in Objects classification.



## Connections between users and interactions
DMH_Stories collected from users
- There is not information about the users/emotions

DMH_91Stories_DEGARI_SemanticAnnotator, DMH_213Stories_DEGARI_SemanticAnnotator, DMH_Narratives 123 stories - DEGARI Classification
- There is information about the emotions, but there is not any attribute(s) identifying the users.

DMH - Pop_up_VR_Museum - Written Stories
- The users are identified with the attribute **Contributor name**, but they cannot be linked to any emotion database.
