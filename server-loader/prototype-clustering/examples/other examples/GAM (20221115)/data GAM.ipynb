## Connections between datasets

GAMGame_stories_RN_UNITO (RN experiment) (https://spice.kmi.open.ac.uk/dataset/details/118)
- **IdArtefact** links to **@id** in GAM_Catalogue_plus
- **userName** links to **userName** in GAMGame_users_RN_UNITO
    
Gam_Game_Story_Definitions (https://spice.kmi.open.ac.uk/dataset/details/58)
- **artworkId** links to **_id** in GAM_Catalogue_plus
- **authorUsername** links to **userName** in GAMGame_users_RN_UNITO or **authorUsername** in  GAMGame - information on registered users

```python
artworks_df2 = artworks.copy()
artworks_df2['_id'] = artworks_df2['_id'].astype(str)

rn_interactions_df2 = rn_interactions.copy() 
rn_interactions_df2['IdArtefact'] = rn_interactions_df2['IdArtefact'].astype(str)
```

```python
rn_interactions_merged_outer = pd.merge(artworks_df2, rn_interactions_df2, left_on='_id', right_on='IdArtefact', how='outer')
rn_interactions_merged_outer
```

```python
# Artworks not belonging to GAM_Catalogue_Plus with which the users interact in RN_stories
artworksNotInCataloguePlus = rn_interactions_merged_outer.loc[rn_interactions_merged_outer['_id_x'] != rn_interactions_merged_outer['IdArtefact']]
artworksNotInCataloguePlus = artworksNotInCataloguePlus.dropna(subset=['IdArtefact'])
artworksNotInCataloguePlus['IdArtefact'] = artworksNotInCataloguePlus['IdArtefact'].astype(str)

artworksNotInCatalogue = pd.unique(artworksNotInCataloguePlus['IdArtefact'])
print(len(artworksNotInCatalogue))
print(artworksNotInCatalogue)
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Interactions with an artwork from GAM_Catalogue_plus (https://spice.kmi.open.ac.uk/dataset/details/113)
#--------------------------------------------------------------------------------------------------------------------------
```

```python
rn_interactions_common = pd.merge(artworks_df2, rn_interactions_df2, left_on='_id', right_on='IdArtefact', how='inner')
rn_interactions_common
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Interactions without proper interaction data
#--------------------------------------------------------------------------------------------------------------------------
```

```python
# Interactions without emotions
result = rn_interactions_common.loc[rn_interactions_common['emotions'].str.len() == 0 ]
result
```

```python
# Interactions without sentiments
result = rn_interactions_common.loc[rn_interactions_common['sentiment'].str.len() == 0 ]
result
```

```python
# Interactions without plutchik emotions
emptyPlutchikEmotions = {
            "anger": 0,
            "anticipation": 0,
            "disgust": 0,
            "fear": 0,
            "joy": 0,
            "sadness": 0,
            "surprise": 0,
            "trust": 0
        }

result = rn_interactions_common.loc[rn_interactions_common['plutchik_emotions'] == emptyPlutchikEmotions]
result[['plutchik_emotions']]
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Combine interactions with user data
#--------------------------------------------------------------------------------------------------------------------------
```

```python
interactions_grouped = rn_interactions_common.groupby("userName").agg(list)
interactions_grouped
```

```python
rn_user_interactions = pd.merge(rn_users, interactions_grouped, on='userName', how='right')
rn_user_interactions.dropna(subset = ['IdArtefact'], inplace = True)
rn_user_interactions
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Users without interactions with valid interaction data
#--------------------------------------------------------------------------------------------------------------------------
```

```python
# users with emotion data
column = 'emotions'
df = rn_user_interactions.explode(column)
df = df.loc[ df[column].str.len() != 0 ]
df = df.groupby("userName").agg(list)
print(len(df))
df[[column]]
```

```python
# users with sentiment data
column = 'sentiment'
df = rn_user_interactions.explode(column)
df = df.loc[ df[column].str.len() != 0 ]
df = df.groupby("userName").agg(list)
print(len(df))
df[[column]]
```

```python
# users with plutchik emotion data
emptyPlutchikEmotions = {
            "anger": 0,
            "anticipation": 0,
            "disgust": 0,
            "fear": 0,
            "joy": 0,
            "sadness": 0,
            "surprise": 0,
            "trust": 0
        }

column = 'plutchik_emotions'
df = rn_user_interactions.explode(column)
df = df.loc[ df[column] != emptyPlutchikEmotions ]
df = df.groupby("userName").agg(list)
df[[column]]
```

```python
# users with emotion or plutchik emotion data
emptyPlutchikEmotions = {
            "anger": 0,
            "anticipation": 0,
            "disgust": 0,
            "fear": 0,
            "joy": 0,
            "sadness": 0,
            "surprise": 0,
            "trust": 0
        }

column = 'plutchik_emotions'
column2 = 'emotions'
df = rn_user_interactions.explode([column,column2])
df = df.loc[ ((df[column] != emptyPlutchikEmotions) | (df[column2].str.len() != 0)) ]
df = df.groupby("userName").agg(list)
print(len(df))
df[[column, column2]]
```
