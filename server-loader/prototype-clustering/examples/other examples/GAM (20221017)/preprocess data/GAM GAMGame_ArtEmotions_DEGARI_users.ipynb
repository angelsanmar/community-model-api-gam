```python
annotated_stories_df = pd.read_json(dataLoader.fileRoute('data/new-annotated-stories.json'))
annotated_stories_df
```

```python
# Prepare interaction data for clustering
# Explode parts array
df = annotated_stories_df.explode('parts')
df = df.reset_index()
```

```python
# Get each column in parts dict

# Get artworkId, emotion, sentiments
df2_a = pd.json_normalize(df["parts"], max_level = 0)
df2_a
```

```python

# Get itMakesMeThinkAbout attribute which is needed for the visualization
df2_b = pd.json_normalize(df2_a["multimediaData"], max_level = 0)
df2_b = pd.json_normalize(df2_b["answersToTemplates"], max_level = 0)
df2_b = df2_b.rename(columns = {'itMakesMeThinkAbout':'itMakesMeThinkAbout'})

df2_b
```

```python
# Interactions user-artwork dataframe
df2 = pd.concat([df2_a,df2_b], axis=1, join='inner')
df2
```

```python
# Set them as new columns
df3 = pd.concat([df,df2], axis=1, join='inner')
df3
```

```python
df3[['emotions']]
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Remove interactions without a valid artwork
#--------------------------------------------------------------------------------------------------------------------------
        
artworksFilename = DataLoader().fileRoute("data/GAM_Catalogue_plus.json")
artworks = pd.read_json(artworksFilename)
artworks['artworkId'] = artworks['_id'].astype(str)
df4 = pd.merge(df3,artworks, on='artworkId', how='right')

print("df4 after merging with artworks")
print(df4)
print("\n\n\n")

df3 = df4
df3
```

```python
# Set default values
"""
values = {'emotions':  {}}
df3 = df3.fillna(value=values)
"""
df3['emotions'] = df3['emotions'].apply(lambda x: {} if x != x else x)
df3[['emotions']]
```

```python
# Set default values
df4 = df3.groupby("authorUsername").agg(list)
df4 = df4.reset_index()
df5 = df4.rename(columns = {'authorUsername':'userName'})

# Combine with user data
filename = dataLoader.fileRoute('data/GAMGame_users_RN_UNITO.json')
users = pd.read_json(filename)
user_interactions = pd.merge(df5, users, on='userName', how='left')

# Set default values
values = {'RelationshipWithArt': '', 'RelationshipWithMuseum': ''}
user_interactions = user_interactions.fillna(value=values)

user_interactions
```

```python
result = user_interactions.to_json(orient='records')
parsed = json.loads(result)
parsed
```
