```python
#--------------------------------------------------------------------------------------------------------------------------
#    Remove interactions without a valid artwork
#--------------------------------------------------------------------------------------------------------------------------
```

```python
print(pd.unique(interactions['IdArtefact']))
print(len(pd.unique(interactions['IdArtefact'])))
```

```python

df = interactions.copy()
df['IdArtefactString'] = interactions['IdArtefact']#.astype(str)
artworks['IdArtefactString'] = artworks['_id']#.astype(str)
```

```python
df[['IdArtefactString']]
```

```python
interactions2 = pd.merge(df,artworks, on='IdArtefactString', how='right')
interactions2.dropna(subset = ['IdArtefact'], inplace = True)
interactions2['IdArtefact'] = interactions2['IdArtefact'].astype(int)
interactions2
```

```python
print(pd.unique(interactions2['IdArtefact']))
print(len(pd.unique(interactions2['IdArtefact'])))
```

```python
#--------------------------------------------------------------------------------------------------------------------------
#    Merge with user data (to obtain demographics)
#--------------------------------------------------------------------------------------------------------------------------
```

```python
interactions2
```

```python
# First group interaction contributions by userName
interactions_grouped = interactions2.groupby("userName").agg(list)
interactions_grouped
```

```python
# https://stackoverflow.com/questions/18792918/combine-two-pandas-data-frames-join-on-a-common-column
# Now merge with user data
user_interactions = pd.merge(users, interactions_grouped, on='userName', how='right')
user_interactions.dropna(subset = ['IdArtefact'], inplace = True)
user_interactions
```

```python
user_interactions.columns
```

```python
# https://stackoverflow.com/questions/50404559/python-error-typeerror-object-of-type-timestamp-is-not-json-serializable
str1 = '_timestamp'
options1 = ['x','y']
options2 = ['day','month','year','hour','minute','second']
columns = []
for option1 in options1:
    columns.append(str1 + '_' + option1)
    for option2 in options2:
        columns.append(str1 + '_' + option2 + '_' + option1)
columns.append('TimeStamps')
columns

```

```python
user_interactions_save = user_interactions.copy()
user_interactions_save[columns] = user_interactions_save[columns].astype(str)
#user_interactions_save.drop(columns, axis = 1)
user_interactions_save
```

```python
# Change NaN values for demographics data (RelationshipWithArt, RelationshipWithMuseum)
values = {'RelationshipWithArt':  'unknown', 'RelationshipWithMuseum': 'unknown'}
user_interactions_save2 = user_interactions_save.fillna(value=values)
user_interactions_save2
```

```python
with open("data/generated/GAM user_interactions mergedArtworks.json", "w") as outfile:
    json.dump(user_interactions_save2.to_dict('records'), outfile, indent=4)
```

```python
# Remove anonymous users
user_interactions_save3 = user_interactions_save2.loc[user_interactions_save2['userName'] != 'anonymous']
user_interactions_save3
```

```python
with open("data/generated/GAM user_interactions mergedArtworks (not anonymous).json", "w") as outfile:
    json.dump(user_interactions_save3.to_dict('records'), outfile, indent=4)
```
